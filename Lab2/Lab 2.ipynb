# Part 1

## 1.
$$x = x_0 + v_0t + \frac{1}{2}at^2, 4000 = \frac{1}{2}gt^2, t = \sqrt{\frac{2 \cdot 4000}{g}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as si
gravitational_constant = 6.6743e-11
mass_earth = 5.972e24
radius_earth = 6378100
rotation_rate = 7.272e-5
mass_moon = 7.35e22
radius_moon = 1738100

# calculate g via law of gravitation
g = gravitational_constant * mass_earth / (radius_earth ** 2)
analytic = np.sqrt(2 * 4000 / g)
print(f"Freefall time (quick calculation): {analytic} seconds")
```

## 2.
$$\frac{dy}{dt} = v, \frac{dv}{dt} = -g$$

## 3.

```python
# s = [dydt, dvdt]
def d(t, s):
    return [s[1], -g]
t = np.linspace(0, 100, 1000)
def event(t, s):
    return s[0] + 4000
event.terminal = True;
solution = si.solve_ivp(d, (0, 100), [0, 0], t_eval=t, events=event)
fig, ax = plt.subplots()
ax.plot(solution.t, solution.y[0], label="position")
ax.hlines(-4000, 0, 30, color="black", linestyle="--", label="bottom of shaft")
ax.set_xlabel("Time (sec)")
ax.set_ylabel("Position relative to Earth's surface (m)")
twinx = ax.twinx()
twinx.plot(solution.t, solution.y[1], label="velocity", color="orange")
twinx.set_ylabel("velocity (m/s)")
fig.suptitle("Figure 1: Position and velocity vs. time")
fig.legend();
```

```python
numerical = solution.t_events[0][0]
print(f"Time when mass hits bottom: {numerical} seconds")
print(f"Analytic - numerical: {analytic - numerical}")
print("np.isclose():", np.isclose(analytic, numerical))
```

The difference between the times obtained analytically and numerically is insignificantly small.

# Part 2

## 1.
$$\frac{dy}{dt} = v, \frac{dv}{dt} = -\frac{r}{R_\oplus}g$$

```python
def linear_gravity(r):
    return -(r/radius_earth) * g
def d(t, s):
    return [s[1], linear_gravity(s[0])]
def event(t, s):
    return s[0] - radius_earth + 4000    # We are measuring position as distance from center of Earth now
event.terminal = True;
solution = si.solve_ivp(d, (0, 100), [radius_earth, 0], t_eval=t, events=event)
fig, ax = plt.subplots()
ax.plot(solution.t, solution.y[0], label="position")
ax.hlines(radius_earth - 4000, 0, 30, color="black", linestyle="--", label="bottom of shaft")
ax.legend()
ax.set_xlabel("Time (sec)")
ax.set_ylabel("Distance from center of Earth (m)")
fig.suptitle("Figure 2: Distance above center of Earth vs. time (linear gravity)")
fig.tight_layout()
```

## 2.

```python
linear_gravity_time = solution.t_events[0][0]
print(f"Time when mass hits bottom: {linear_gravity_time} seconds")
print(f"Linear-gravity fall time - Constant-gravity fall time: {linear_gravity_time - analytic}")
```

It takes longer in the linear-gravity case for the mass to hit the bottom of the shaft, as the mass's acceleration decreases over time in the linear-gravity case, but the difference between the constant-gravity and linear-gravity cases is around a thousandth of a second.

## 3.

```python
def d(t, s, alpha, gamma):
    # Linear gravity + drag
    return [s[1], linear_gravity(s[0]) + alpha * np.abs(s[1]) ** gamma]
def event(t, s, alpha, gamma):
    return s[0] - radius_earth + 4000
event.terminal=True
solution = si.solve_ivp(d, (0, 100), [radius_earth, 0], t_eval=t, args=[0.00395, 2], events=event)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(solution.t, solution.y[0], label="position")
ax1.hlines(radius_earth - 4000, 0, 100, color="black", linestyle="--", label="bottom of shaft")
ax1.set_title("Position vs. time")
ax1.set_xlabel("Time (sec)")
ax1.set_ylabel("Distance from center of Earth (m)")
ax1.legend()
ax2.plot(solution.t, solution.y[1], label="velocity")
ax2.hlines(-50, 0, 100, color="black", linestyle="--", label="terminal velocity")
ax2.set_title("Velocity vs. time")
ax2.set_xlabel("Time (sec)")
ax2.set_ylabel("Velocity (m/s)")
ax2.legend()
fig.suptitle("Fig 2.5: Position and velocity vs. time (linear gravity + drag)")
fig.tight_layout()
```

```python
linear_gravity_with_drag_time = solution.t_events[0][0]
print(f"Fall time: {linear_gravity_with_drag_time} seconds")
print(f"Linear-gravity with drag fall time - Constant-gravity fall time: {linear_gravity_with_drag_time - analytic} seconds")
print(f"Linear-gravity with drag fall time - Linear-gravity without drag: {linear_gravity_with_drag_time - linear_gravity_time} seconds")
```

Including drag significantly increases the fall time, much more than incorporating linear gravity.

# Part 3

## 1.
$$\frac{dx}{dt} = v_x, \frac{dv_x}{dt} = 2\Omega v_y$$
$$\frac{dy}{dt} = v_y, \frac{dv_y}{dt} = -2\Omega v_x$$

## 2.

```python
# [dx, dvx, dy, dvy]
# This time, y is the depth in meters, because that makes the plot make more sense.
def d(t, s, alpha, gamma):
    # Linear gravity + drag + Coriolis
    return [s[1], 2 * rotation_rate * np.abs(s[3]) - alpha * np.abs(s[1]) ** gamma,
            -s[3], linear_gravity(radius_earth - s[2]) + alpha * np.abs(s[3]) ** gamma + 2 * rotation_rate * s[1]]

# New event function because we switched to y measuring the depth in meters
def new_event(t, s, alpha, gamma):
    return s[2] - 4000
new_event.terminal = True

solution = si.solve_ivp(d, (0, 100), [0, 0, 0, 0], t_eval=t, args=[0, 0], events=new_event)
fig, ax = plt.subplots()
ax.plot(solution.y[2], solution.y[0], label="position")
ax.hlines(2.5, 0, 4000, linestyle="--", color="black", label="side of shaft")
ax.set_title("Figure 3: Transverse position vs. depth (Linear gravity, no drag)")
ax.set_xlabel("Depth (m)")
ax.set_ylabel("Transverse position (m)")
ax.legend()
fig.tight_layout()
```

## 3.
If the mineshaft is 5m wide and the mass is dropped from the center, the mass will indeed bump into the wall. It moves just over 5 meters in the transverse direction, which is more than the radius of the hole (2.5m).

## 4.

```python
solution = si.solve_ivp(d, (0, 100), [0, 0, 0, 0], t_eval=t, args=[0.00395, 2], events=new_event)
fig, ax = plt.subplots()
ax.plot(solution.y[2], solution.y[0], label="position")
ax.hlines(2.5, 0, 4000, linestyle="--", color="black", label="side of shaft")
ax.set_title("Figure 3.5: Transverse position vs. depth (Linear gravity + drag + Coriolis)")
ax.set_xlabel("Depth (m)")
ax.set_ylabel("Transverse position (m)")
ax.legend()
fig.tight_layout()
```

```python
def hit_wall(t, s, alpha, gamma):
    return s[0] - 2.5

solution = si.solve_ivp(d, (0, 100), [0, 0, 0, 0], t_eval=t, args=[0.00395, 2], events=hit_wall)
print(f"Depth when mass hits wall: {solution.y_events[0][0][2]} meters")
print(f"Time when mass hits wall: {solution.t_events[0][0]} seconds")
```

Drag makes a very big difference: The mass moves almost 5 times farther in the transverse direction than in the no-drag scenario. This is because the vleocity in the y direction is significantly larger than in the x direction, so the mass experiences a much larger drag in the y direction; in other words, the x velocity as a function of time is quite similar to the no-drag scenario, but the y velocity as a function of time is significantly less than in the no-drag scenario. Therefore, the mass moves farther in the transverse direction with drag, as it takes significantly longer for the mass to reach the bottom.

```python
# linear_gravity will flip signs when our y position does
def d(t, s):
    return [s[1], linear_gravity(s[0])]
t = np.linspace(0, 10000, 1000)
solution = si.solve_ivp(d, (0, 10000), [radius_earth, 0], t_eval=t)
fig, ax = plt.subplots()
ax.plot(solution.t, solution.y[0])
ax.set_xlabel("Time (sec)")
ax.set_ylabel("Position (m)")
fig.suptitle("Figure 4: Position vs. time (infinitely deep mine, linear gravity, no drag/Coriolis force)")
fig.tight_layout()
```

For an infinitely deep mine, we actually get simple harmonic motion, as the force on the mass is linearly proportional to its position, as if it were attached to a spring.

```python
# Happens when y = 0.
def reach_center(t, s):
    return s[0]
# Happens when velocity is 0.
def reach_other_side(t, s):
    return s[1]
solution = si.solve_ivp(d, (0, 10000), [radius_earth, 0], t_eval=t, events=[reach_center, reach_other_side])
print(f"Time to reach other side: {solution.t_events[1][1]} seconds")
print(f"Time to reach center: {solution.t_events[0][0]} seconds")
print(f"Velocity at center: {solution.y_events[0][0][1]} m/s")
```

The time it would take for a mass to fall from one side of the Earth to the other is just over four times faster than the time it would take for a mass to orbit around the Earth.

# Part 5

## 1.

```python
def density(rho, r, n):
    return rho * (1 - r**2 / radius_earth**2)**n
r = np.linspace(1, radius_earth, 1000)
fig, ax = plt.subplots()
for n in [0, 1, 2, 9]:
    ax.plot(r, density(1, r, n), label=f"n = {n}")
ax.legend()
ax.set_xlabel("Radius (m)")
ax.set_ylabel("Density (kg/m$^3$)")
fig.suptitle("Figure 5: Normalized density profile vs. radius")
fig.tight_layout()
```

## 2.
$$M_\oplus = 4\pi\rho_n\int_0^{R_\oplus}\Big(1 - \frac{r^2}{R_\oplus^2}\Big)^nr^2dr, \rho_n = \frac{M_\oplus}{4\pi\int_0^{R_\oplus}\Big(1 - \frac{r^2}{R_\oplus^2}\Big)^nr^2dr}$$

```python
def mass(rho, r, n):
    return 4 * np.pi * si.quad(lambda radius : rho * (1 - radius**2 / radius_earth**2)**n * radius**2, 0, r)[0]

def linear_gravity_density(rho, r, n):
    return gravitational_constant * mass(rho, r, n) / r**2

fig, ax = plt.subplots()
for n in [0, 1, 2, 9]:
    rho_normalized = mass_earth / (4 * np.pi * si.quad(lambda radius : (1 - radius**2 / radius_earth**2)**n * radius**2, 0, radius_earth)[0])
    values = []
    for radius in r:
        values.append(linear_gravity_density(rho_normalized, radius, n))
    ax.plot(r, values, label=f"n = {n}")

ax.legend()
ax.set_xlabel("Radius (m)")
ax.set_ylabel("Force profile (N)")
fig.suptitle("Figure 6: Force profile vs. radius (1kg test mass)")
fig.tight_layout()
```

## 3.

```python
def d(t, s, rho, n):
    return [s[1], -linear_gravity_density(rho, s[0], n)]

fig, (a1, a2) = plt.subplots(1, 2, figsize=(12, 5))
t = np.linspace(0, 3500, 1000)
for n in [0, 1, 2, 9]:
    rho_normalized = mass_earth / (4 * np.pi * si.quad(lambda radius : (1 - radius**2 / radius_earth**2)**n * radius**2, 0, radius_earth)[0])
    solution = si.solve_ivp(d, (0, 3500), [radius_earth, 0], args=[rho_normalized, n], t_eval=t)
    a1.plot(solution.t, solution.y[0], label=f"n = {n}")
    a2.plot(solution.t, solution.y[1], label=f"n = {n}")

a1.legend()
a1.set_xlabel("Time (sec)")
a1.set_ylabel("Position (m)")
a1.set_title("Position vs. time")
a2.legend()
a2.set_xlabel("Time (sec)")
a2.set_ylabel("Velocity (m/s)")
a2.set_title("Velocity vs. time")
fig.suptitle("Figure 7: Position and velocity vs. time (non-uniform density)")
fig.tight_layout()
```

## 4.

```python
def reach_center(t, s, rho, n):
    return s[0]
for n in [0, 1, 2, 9]:
    rho_normalized = mass_earth / (4 * np.pi * si.quad(lambda radius : (1 - radius**2 / radius_earth**2)**n * radius**2, 0, radius_earth)[0])
    solution = si.solve_ivp(d, (0, 3500), [radius_earth, 0], args=[rho_normalized, n], t_eval=t, events=reach_center)
    print(f"Time to reach center for n = {n}: {solution.t_events[0][0]} seconds; velocity: {solution.y_events[0][0][1]} m/s")
```

# Part 6

## 1.

```python
g_moon = gravitational_constant * mass_moon / (radius_moon ** 2)

def linear_gravity_moon(r):
    return (r/radius_moon) * g_moon

def d(t, s):
    return [s[1], -linear_gravity_moon(s[0])]

def reach_center(t, s):
    return s[0]

t = np.linspace(0, 10000, 1000)
solution = si.solve_ivp(d, (0, 10000), [radius_moon, 0], t_eval=t, events=reach_center)
print(f"Time to reach center of Moon: {solution.t_events[0][0]} seconds")
```

## 2.

```python
print(f"Density of Moon: {mass_moon / ((4/3) * np.pi * radius_moon**3)} kg/m^3")
print(f"Density of Earth: {mass_earth / ((4/3) * np.pi * radius_earth**3)} kg/m^3")
```

Assuming a constant density, the mass of a planetary body is equal to $\rho\frac{4}{3}\pi r^3$. Then, the acceleration of a mass at a radius $r$, for $|r|$ less than the radius of the planetary body, is equal to $G\rho\frac{4}{3}\pi r$, giving us the differential equation:
$$\frac{d^2x}{dt} = -G\rho\frac{4}{3}\pi x$$
The solution to this equation is obviously some sinusoidal function with angular frequency $\sqrt{G\rho\frac{4}{3}\pi}$. The time it takes for the mass to reach the other side of the planetary body would then be $\frac{\pi}{\sqrt{G\rho\frac{4}{3}\pi}} = \frac{\sqrt\pi}{\sqrt{G\rho\frac{4}{3}}}$ seconds.
